### Function for pre-processing the textual data

```python
def preprocess_text(df):
    
    #Strip Punctation
    df[0]=df[0].apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '', x))
#     df[0] = df[0].apply(lambda x: re.sub(r'[^\w\s]','',x))


    #Remove Non-words
    df[0] = df[0].apply(lambda x: re.sub(r'[^a-zA-Z\s]','',x))

    #To lower
    df[0] = df[0].apply(lambda x: x.lower())

    #Remove multiple letter repating words
    df[0] = df[0].apply(lambda x: re.sub(r'([a-z])\1{2,}[\s|\w]*','',x)) 

    return df
```

```python
# apply pre-processing
df_lex = preprocess_text(df_lex)
```

```python
# use spacy for stopword and lemmatization
nlp = spacy.load('en_core_web_sm',disable=['parser', 'ner'])
```

```python
# Lemmatization with stopwords removal
df_lex['lemmatized']=df_lex[0].apply(lambda x: ' '.join([token.lemma_ for token in list(nlp(x)) if (token.is_stop==False)]))
```

```python
df_lex["len_lemm"] = df_lex["lemmatized"].apply(len)
```

```python
def generate_wordcloud(data,title=None):
    wc = WordCloud(width=400, height=330, max_words=150,colormap="Dark2").generate(" ".join(df_lex[0]))
    plt.figure(figsize=(10,8))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
    if title is None:
        plt.title("Lex' Sentence")
    plt.title(title)
#     plt.title('\n'.join(wrap(title,60)),fontsize=13)
    plt.show()
```
